## 3.2.1 The mpg data frame

```
# Read in and display the data
url = "https://github.com/byuidatascience/data4python4ds/raw/master/data-raw/mpg/mpg.csv"
mpg = pd.read_csv(url)
mpg
```

## 3.2.2 Creating an Altair plot

```
displ_vs_hwy = (alt.Chart(mpg)
  .encode(
    x='displ', 
    y='hwy')
  .mark_circle()
)

displ_vs_hwy
```
